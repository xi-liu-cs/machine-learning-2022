```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spambase/spambase.data')
df
```

```python
y = df.iloc[:, -1:] # last column in spambase.data is the label (y) for training the model, read last column from df
y = np.array(y).ravel()
x = df.iloc[: , :-1] # before train the model, remove label from data, remove last column from df
x = x.copy()
x = np.array(x) # transfer df to array
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0) # split dataset into train and test dataset paired with its label
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python

from sklearn.svm import SVC

def svm(C, kernel):
    classifier = SVC(C = C, kernel = kernel, degree = 2, gamma = 'auto') # set up kernel and hyper-parameter (C, gamma) for the model
    classifier.fit(x_train, y_train) # train the model
    train_score = classifier.score(x_train, y_train) # inputs are train dataset and train label
    test_score = classifier.score(x_test, y_test) # inputs are test dataset and test label
    return train_score, test_score
```

```python
kernel_list = ['linear', 'rbf', 'poly']
C_list = [0.01, 0.1, 1, 10, 100, 1000, 10000]

for kernel in kernel_list:
    for C in C_list:
        train_score, test_score = svm(C, kernel)
        print('C:', C)
        print('kernel:', kernel)
        print('train accuracy: ', train_score)
        print('test accuracy: ', test_score, '\n')
```
